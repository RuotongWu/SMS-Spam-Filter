```python
import numpy as np
import pickle
import import_ipynb
import os
import pandas as pd
from data_helper import data_loader
import tensorflow as tf
```

```python
class lstm_classifier:
    
    def __init__(self,file_name):
        self.__load_processed_data(file_name)
        #build NN network with lstm
        self.__model = tf.keras.Sequential([
            tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128)),
            tf.keras.layers.Dense(128,activation='relu'),
            tf.keras.layers.Dense(1,activation='sigmoid')
        ])
        
        self.__model.compile(loss = 'binary_crossentropy',
                          optimizer='adam',
                          metrics=['accuracy'])
        #fit the model
        history = self.__model.fit(self.__train_message,self.__train_tag, epochs=10)
        
    
    def __load_processed_data(self, file_name,split_ratio = 0.9, shuffle = False, encoding = True):
        current_path = os.getcwd()
        pickle_data_folder = current_path+ r"/data/processed"
        file_path = pickle_data_folder + "//" + file_name
        with open(file_path,'rb') as f:
            self.__data = pickle.load(f) 
        self.__train_message, self.__train_tag, self.__test_message, self.__test_tag = self.__data.data_split()
        
            
    def evaluation(self):
        return self.__model.evaluate(self.__test_message, self.__test_tag)
    
    def prediction(self):
        pre = self.__model.predict(self.__test_message)
        pre=list(map(lambda x:int(1.0-x<=x-0.),pre))
        count = 0
        for p,r in zip(pre, self.__test_tag):
            if p == r:
                count += 1
            
        return self.__model.predict(self.__test_message),count/len(pre)
        
    
```
